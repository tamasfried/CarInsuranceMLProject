# Plotting w/o fitting to a model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/car_insurance.csv")

X = df['Driver Experience']
y = df['Insurance Premium ($)']
```

```python
plt.figure(figsize=(10,7.5))
plt.scatter(X, y, color='blue', label='Training Data')
plt.xlabel("Driver Exp")
plt.ylabel("Insurance Premium ($)")
plt.title(f"Driver Exp vs Insurance Premiums")
plt.grid(True)
plt.legend()
plt.show()
```
